## Parse Data and Extract Legend

### Subtask:
Parse the JSON data from `json_data_str` to extract all `text_blocks`. Then, extract and analyze the legend (text block with id '28') to understand the terminology for sand ('piaski', 'p') and gravel ('żwiry', 'ż').


**Reasoning**:
The subtask requires parsing the `json_data_str` into a Python dictionary, extracting the `text_blocks`, and then isolating the legend text from the block with ID '28'. Although `eval()` was used previously, `json.loads()` is the standard and safer way to parse JSON strings.



```python
import json

# Convert the string to a Python dictionary using json.loads()
data = json.loads(json_data_str)

# Extract the list of text blocks
text_blocks = data.get('text_blocks', [])

# Iterate through text_blocks to find the dictionary with id '28'
legend_text = None
for block in text_blocks:
    if block.get('id') == '28':
        legend_text = block.get('text_content')
        break

# Print the extracted legend to verify
print("Extracted Legend (ID '28'):\n", legend_text)
```

**Reasoning**:
The `JSONDecodeError` occurred because `json.loads()` requires strict JSON syntax (double quotes for keys and string values), but `json_data_str` is formatted as a Python dictionary literal using single quotes. To fix this, I will use `ast.literal_eval()` which can safely parse Python literal structures from a string, achieving the intended dictionary conversion.



```python
import json
import ast

# Convert the string to a Python dictionary using ast.literal_eval()
# json.loads() expects strict JSON (double quotes), but the string uses single quotes for keys/strings.
data = ast.literal_eval(json_data_str)

# Extract the list of text blocks
text_blocks = data.get('text_blocks', [])

# Iterate through text_blocks to find the dictionary with id '28'
legend_text = None
for block in text_blocks:
    if block.get('id') == '28':
        legend_text = block.get('text_content')
        break

# Print the extracted legend to verify
print("Extracted Legend (ID '28'):\n", legend_text)
```

**Reasoning**:
Now that the `legend_text` has been successfully extracted, the next step is to programmatically analyze this text to define the specific terminology for 'sand' ('piaski', 'p') and 'gravel' ('żwiry', 'ż') and their variants. This involves parsing the descriptive part of the legend to create keyword lists that will be used to identify relevant geological layers.



```python
import re

# Define keywords for sand and gravel based on the legend analysis
# 'piaski' and 'piasków' for sand, 'żwiry' and 'żwirów' for gravel.
# For structural explanations, 'p' is sand, 'ż' is gravel.
# Variants of 'p' are 'pd', 'ps', 'pg', 'ppy'.

sand_keywords = ['piaski', 'piasków', 'p', 'pd', 'ps', 'pg', 'ppy']
gravel_keywords = ['żwiry', 'żwirów', 'ż']

# Additionally, parse the legend to map layer codes (A, B0, B1, etc.) to their descriptions
layer_descriptions = {}

# Regular expression to capture layer codes and their descriptions
# Looks for 'CODE - description' pattern
pattern = r'([A-Z0-9]+) - ([^;]+?(?:piaski|żwiry|piasków|żwirów)[^;]+?)(?:$|;|,|\n)'

# Split the legend text into main parts and structural explanations
main_legend, structural_legend = legend_text.split('Dodatkowe objaśnienia strukturalne:')

# Extract layer descriptions from the main legend part
matches = re.findall(pattern, main_legend, re.IGNORECASE)

for code, desc in matches:
    layer_descriptions[code.strip()] = desc.strip()

print("Sand Keywords:", sand_keywords)
print("Gravel Keywords:", gravel_keywords)
print("Layer Descriptions:", layer_descriptions)
```

**Reasoning**:
With the sand and gravel keywords and layer descriptions defined, the next step is to iterate through all `text_blocks` to identify geological layers containing sand or gravel. This involves checking both the layer codes (like A, B1, C) and shorthand descriptions (like ps+ż, pd) against the extracted terminology.



```python
sand_layers = []
gravel_layers = []

# Helper function to check if a text contains sand keywords
def contains_sand(text):
    text_lower = text.lower()
    for keyword in sand_keywords:
        if keyword in text_lower:
            return True
    return False

# Helper function to check if a text contains gravel keywords
def contains_gravel(text):
    text_lower = text.lower()
    for keyword in gravel_keywords:
        if keyword in text_lower:
            return True
    return False

# Iterate through all text blocks to identify sand and gravel layers
for block in text_blocks:
    block_id = block.get('id')
    text_content = block.get('text_content', '')
    bounding_box = block.get('bounding_box')
    block_type = block.get('block_type')

    # Skip caption blocks that are not relevant layer descriptions
    if block_type == 'caption' and block_id not in layer_descriptions.keys():
        continue

    # Check based on layer codes (A, B1, C, etc.) from the main legend
    if text_content in layer_descriptions:
        description = layer_descriptions[text_content]
        if contains_sand(description):
            sand_layers.append({'id': block_id, 'text_content': text_content + ' (' + description + ')', 'bounding_box': bounding_box})
        if contains_gravel(description):
            gravel_layers.append({'id': block_id, 'text_content': text_content + ' (' + description + ')', 'bounding_box': bounding_box})

    # Check for direct shorthand labels (ps+ż, pd, etc.)
    elif block.get('text_type') == 'label':
        if contains_sand(text_content):
            sand_layers.append({'id': block_id, 'text_content': text_content, 'bounding_box': bounding_box})
        if contains_gravel(text_content):
            gravel_layers.append({'id': block_id, 'text_content': text_content, 'bounding_box': bounding_box})

# Print the summarized sand and gravel layers
print("Identified Sand Layers:")
for layer in sand_layers:
    print(layer)

print("\nIdentified Gravel Layers:")
for layer in gravel_layers:
    print(layer)
```
